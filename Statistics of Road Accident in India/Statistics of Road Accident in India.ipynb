## Asking and Answering Questions
#### Q1: Death Percentage of Fatalities in 2016 and 2017


```python
Total_Death = merged_df['Death in 2016'].sum()

Population_2016 = Road_Accidents.loc[2016]['Population of India (in thousands)']

Ratio = Total_Death/Population_2016

F_Ratio = Ratio * 100

print('In India {:.2f}% of humans are dead in accident during 2016'.format(F_Ratio))
```

```python
Total_Death = merged_df['Death in 2017'].sum()

Population_2016 = Road_Accidents.loc[2017]['Population of India (in thousands)']

Ratio = Total_Death/Population_2016

F_Ratio = Ratio * 100

print('In India {:.2f}% of humans are dead in accident during 2016'.format(F_Ratio))
```

#### Q2: From Which Type of Vehicle having the Death numbers most in 2016 and 2017?


```python
col = Road_Accident_V_Type['2016 - Persons killed']
max_v = col.max()
max_v

index = Road_Accident_V_Type.index
condition = Road_Accident_V_Type['2016 - Persons killed'] == max_v
get_index = index[condition]
in_list = get_index.tolist()

vehicle_types = Road_Accident_V_Type['Vehicle Type']

sel_type = vehicle_type[in_list[0]]

print(f"Most accident happen in 2016 due to {sel_type}")
```

```python
col = Road_Accident_V_Type['2017 - Persons killed']
max_v = col.max()
max_v

index = Road_Accident_V_Type.index
condition = Road_Accident_V_Type['2017 - Persons killed'] == max_v
get_index = index[condition]
in_list = get_index.tolist()

vehicle_types = Road_Accident_V_Type['Vehicle Type']

sel_type = vehicle_type[in_list[0]]

print(f"Most accident happen in 2017 due to {sel_type}")
```

`Conclusion - In both year 2016 and 2017 most of accident happens due to Two Wheelers because Two wheelers have less safety compare to other vehicles`
